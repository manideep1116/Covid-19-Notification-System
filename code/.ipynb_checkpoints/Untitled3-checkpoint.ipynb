```python

#LSTM Implementation on Covid Data

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
import tensorflow as tf

data = pd.read_csv('cases.csv')
MO_cases = data["State"]
County_cases = data["County"]
Zip_cases = data["Zip"]
```

```python
plt.figure()
#plt.plot(data['State'])
#plt.plot(data['County'])
#plt.plot(data['Zip'])
plt.title('Covid Cases in Missouri')
plt.plot(MO_cases, color = 'brown')
plt.xlabel('Days')
plt.ylabel('Daily New Cases')
plt.legend(['MO'], loc='upper left')
plt.show()

plt.figure()
plt.plot(County_cases, color = 'purple')
#plt.plot(data['Zip'])
plt.title('Covid Cases in St.Louis County')
plt.xlabel('Days')
plt.ylabel('Daily New Cases')
plt.legend(['County'], loc='upper left')
plt.show()

plt.figure()
plt.plot(Zip_cases, color = 'crimson')
plt.title('Covid Cases in St.Louis County')
plt.xlabel('Days')
plt.ylabel('Daily New Cases')
plt.legend(['County'], loc='upper left')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_cols = ["State","County","Zip"]
covid_train, covid_test = train_test_split(data, train_size=0.8, test_size=0.2, shuffle=False)
print("Train and Test size", len(covid_train), len(covid_test))
# scale the feature MinMax, build array
x = covid_train.loc[:,train_cols].values
min_max_scaler = MinMaxScaler()
x_train = min_max_scaler.fit_transform(x)
x_test = min_max_scaler.transform(covid_test.loc[:,train_cols])
```

```python
#Build time series format
TIME_STEPS=10
BATCH_SIZE = 32
from tqdm import tqdm
def build_timeseries(mat, y_col_index):
    dim_0 = mat.shape[0] - TIME_STEPS
    dim_1 = mat.shape[1]
    x = np.zeros((dim_0, TIME_STEPS, dim_1))
    y = np.zeros((dim_0,))
    
    for i in tqdm(range(dim_0)):
        x[i] = mat[i:TIME_STEPS+i]
        y[i] = mat[TIME_STEPS+i, y_col_index-1]
    print("length of time-series i/o",x.shape,y.shape)
    return x, y
```

```python
def trim_dataset(mat, batch_size):
    no_of_rows_drop = mat.shape[0]%batch_size
    if(no_of_rows_drop > 0):
        return mat[:-no_of_rows_drop]
    else:
        return mat
```

```python
x_t, y_t = build_timeseries(x_train, 3)
x_t = trim_dataset(x_t, BATCH_SIZE)
y_t = trim_dataset(y_t, BATCH_SIZE)
x_temp, y_temp = build_timeseries(x_test, 3)
x_val, x_test_t = np.split(trim_dataset(x_temp, BATCH_SIZE),2)
y_val, y_test_t = np.split(trim_dataset(y_temp, BATCH_SIZE),2)
```

```python

#Modeling with keras LSTM
import tensorflow.keras
#import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
import tensorflow.keras.optimizers
#from tensorflow.keras import backend
lstm_model = Sequential()
lstm_model.add(LSTM(100, batch_input_shape=(BATCH_SIZE, TIME_STEPS, x_t.shape[2]), dropout=0.0, recurrent_dropout=0.0, stateful=True, kernel_initializer='random_uniform'))
lstm_model.add(Dropout(0.5))
lstm_model.add(Dense(20,activation='relu'))
lstm_model.add(Dense(1,activation='sigmoid'))
optimizer = tensorflow.keras.optimizers.RMSprop(lr=0.001)
lstm_model.compile(loss='mean_squared_error', optimizer=optimizer)
print(lstm_model.summary())
```

```python
from tensorflow.keras import callbacks
import os
#csv_logger = callbacks.CSVLogger(os.path.join("C:\\Desktop\\Academics\\Third Sem\\FE520\\aaa.log"), append=True)

history = lstm_model.fit(x_t, y_t, epochs=100, verbose=2, batch_size=BATCH_SIZE,
                    shuffle=False, validation_data=(trim_dataset(x_val, BATCH_SIZE),
                    trim_dataset(y_val, BATCH_SIZE)))
```

```python
#Prediction
predicted_cases= lstm_model.predict(x_test_t)
```
